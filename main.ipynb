```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("data/train.csv")
testdf = pd.read_csv("data/test.csv")
```

```python
categorical_columns = []
for col in df.columns:
    if(df[col].dtype=="object"):
        categorical_columns.append(col)
df = pd.get_dummies(df,columns=categorical_columns)
df.apply(lambda x: x.fillna(x.mean()),axis=0)

categorical_columns = []
for col in testdf.columns:
    if(testdf[col].dtype=="object"):
        categorical_columns.append(col)
testdf = pd.get_dummies(testdf,columns=categorical_columns)
testdf.apply(lambda x: x.fillna(x.mean()),axis=0)
testdf.head()
```

```python
x_columns = [x for x in testdf.columns if x not in ["Id"]]
X = df[x_columns].values.tolist()
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
X = imp.fit_transform(X)
for i in range(len(X)):
    X[i] = np.array(X[i])
Y = df["SalePrice"].values.tolist()
for i in range(len(Y)):
    Y[i] = np.array([Y[i]])

```

```python
reg = LinearRegression().fit(X, Y)
reg.score(X, Y)
```

```python
test_x_columns = [x for x in testdf.columns if x not in ["Id"]]
testX = testdf[test_x_columns].values.tolist()
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
testX = imp.fit_transform(testX)
for i in range(len(testX)):
    testX[i] = np.array(testX[i])
```

```python
testY = reg.predict(testX)
```

```python
data = {"Id":[],"SalePrice":[]}
for i in range(len(testY)):
    data["Id"].append(i+1)
    data["SalePrice"].append(testY[i][0])
results = pd.DataFrame (data, columns = ['Id','SalePrice'])
results.reset_index(drop=True, inplace=True)

results.head()
```

```python
results.to_csv("submission.csv")
```
